```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#CNPJ 1

```python
cnpj1 = pd.read_csv('/content/cnpjs_1.csv', sep=';')
```

```python
cnpj1.head()
```

```python
cnpj1.describe()
```

```python
cnpj1['data_inicio_atividade'] = pd.to_datetime(cnpj1['data_inicio_atividade'])
cnpj1.set_index('data_inicio_atividade', inplace=True)
cnpj1.resample('Y').size().plot()
plt.show()
```

```python
sns.countplot(x='sigla_uf', data=cnpj1)
plt.show()
```

```python
sns.countplot(x='cnae_fiscal_principal', data=cnpj1)
plt.show()
```
