# Censo Educação Superior

## Base 1

```python
# Base 1 - INEP Censo Educação Superior

arquivo_censo_educ_sup = '/content/drive/MyDrive/dados_inep/censo_da_educacao_superior/MICRODADOS_ED_SUP_IES_2022.CSV'

df_censo_educ_sup = pd.read_csv(arquivo_censo_educ_sup, encoding='latin1', sep=';')
df_censo_educ_sup
```

```python
# Mapear códigos de região para nomes de região
mapeamento_regiao = {
    1: 'Norte',
    2: 'Nordeste',
    3: 'Sudeste',
    4: 'Sul',
    5: 'Centro-Oeste'
}

# Mapear os códigos para os nomes
df_censo_educ_sup['CO_REGIAO_IES'] = df_censo_educ_sup['CO_REGIAO_IES'].map(mapeamento_regiao)
```

```python
# Gráfico de contagem para a coluna NO_REGIAO_IES
plt.figure(figsize=(10, 6))
sns.countplot(x='NO_REGIAO_IES', data=df_censo_educ_sup, palette='Set3')
plt.title('Distribuição das Instituições por Região')
plt.xlabel('Região da Instituição')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de contagem para a coluna CO_REGIAO_IES
plt.figure(figsize=(10, 6))
sns.countplot(x='CO_REGIAO_IES', data=df_censo_educ_sup, palette='Set2')
plt.title('Distribuição das Instituições por Código de Região')
plt.xlabel('Código da Região da Instituição')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de contagem para a coluna NO_UF_IES
plt.figure(figsize=(14, 8))
sns.countplot(x='NO_UF_IES', data=df_censo_educ_sup, palette='Set1')
plt.title('Distribuição das Instituições por Estado')
plt.xlabel('Estado da Instituição')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')  # Rotacionar rótulos do eixo x para melhor legibilidade
plt.show()
```

## Base 2

```python
# Base 2 - INEP Enceja dados alunos provas

arquivo_2 = '/content/drive/MyDrive/dados_inep/enceja/MICRODADOS_ENCCEJA_2022_PPL_NAC.csv'

df_2 = pd.read_csv(arquivo_2, encoding='latin1', sep=';')
df_2
```

```python
df_2.info()
```

```python
df_2.nunique()
```

```python
# Calcular a média das notas de redação por estado
media_notas_redacao_por_estado = df_2.groupby('SG_UF_PROVA')['NU_NOTA_REDACAO'].mean().reset_index()

# Ordenar os estados pela média das notas
media_notas_redacao_por_estado = media_notas_redacao_por_estado.sort_values(by='NU_NOTA_REDACAO', ascending=False)

# Criar um gráfico de barras
plt.figure(figsize=(14, 8))
sns.barplot(x='SG_UF_PROVA', y='NU_NOTA_REDACAO', data=media_notas_redacao_por_estado, palette='viridis')
plt.title('Média das Notas de Redação por Estado')
plt.xlabel('Estado da Prova')
plt.ylabel('Média das Notas de Redação')
plt.show()
```
