# Censo EducaÃ§Ã£o Superior

## Base 1

```python
# Base 1 - INEP Censo EducaÃ§Ã£o Superior

arquivo_censo_educ_sup = '/content/drive/MyDrive/dados_inep/censo_da_educacao_superior/MICRODADOS_ED_SUP_IES_2022.CSV'

df_censo_educ_sup = pd.read_csv(arquivo_censo_educ_sup, encoding='latin1', sep=';')
df_censo_educ_sup
```

```python
# Mapear cÃ³digos de regiÃ£o para nomes de regiÃ£o
mapeamento_regiao = {
    1: 'Norte',
    2: 'Nordeste',
    3: 'Sudeste',
    4: 'Sul',
    5: 'Centro-Oeste'
}

# Mapear os cÃ³digos para os nomes
df_censo_educ_sup['CO_REGIAO_IES'] = df_censo_educ_sup['CO_REGIAO_IES'].map(mapeamento_regiao)
```

```python
# GrÃ¡fico de contagem para a coluna NO_REGIAO_IES
plt.figure(figsize=(10, 6))
sns.countplot(x='NO_REGIAO_IES', data=df_censo_educ_sup, palette='Set3')
plt.title('DistribuiÃ§Ã£o das InstituiÃ§Ãµes por RegiÃ£o')
plt.xlabel('RegiÃ£o da InstituiÃ§Ã£o')
plt.ylabel('Contagem')
plt.show()
```

```python
# GrÃ¡fico de contagem para a coluna CO_REGIAO_IES
plt.figure(figsize=(10, 6))
sns.countplot(x='CO_REGIAO_IES', data=df_censo_educ_sup, palette='Set2')
plt.title('DistribuiÃ§Ã£o das InstituiÃ§Ãµes por CÃ³digo de RegiÃ£o')
plt.xlabel('CÃ³digo da RegiÃ£o da InstituiÃ§Ã£o')
plt.ylabel('Contagem')
plt.show()
```

```python
# GrÃ¡fico de contagem para a coluna NO_UF_IES
plt.figure(figsize=(14, 8))
sns.countplot(x='NO_UF_IES', data=df_censo_educ_sup, palette='Set1')
plt.title('DistribuiÃ§Ã£o das InstituiÃ§Ãµes por Estado')
plt.xlabel('Estado da InstituiÃ§Ã£o')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right')  # Rotacionar rÃ³tulos do eixo x para melhor legibilidade
plt.show()
```

```python
# Base 2 - INEP Enceja dados alunos provas

arquivo_2 = '/content/drive/MyDrive/dados_inep/enceja/MICRODADOS_ENCCEJA_2022_PPL_NAC.csv'

df_2 = pd.read_csv(arquivo_2, encoding='latin1', sep=';')
df_2
```

```python
df_2.info()
```

```python
df_2.nunique()
```

```python
# Calcular a mÃ©dia das notas de redaÃ§Ã£o por estado
media_notas_redacao_por_estado = df_2.groupby('SG_UF_PROVA')['NU_NOTA_REDACAO'].mean().reset_index()

# Ordenar os estados pela mÃ©dia das notas
media_notas_redacao_por_estado = media_notas_redacao_por_estado.sort_values(by='NU_NOTA_REDACAO', ascending=False)

# Criar um grÃ¡fico de barras
plt.figure(figsize=(14, 8))
sns.barplot(x='SG_UF_PROVA', y='NU_NOTA_REDACAO', data=media_notas_redacao_por_estado, palette='viridis')
plt.title('MÃ©dia das Notas de RedaÃ§Ã£o por Estado')
plt.xlabel('Estado da Prova')
plt.ylabel('MÃ©dia das Notas de RedaÃ§Ã£o')
plt.show()
```
