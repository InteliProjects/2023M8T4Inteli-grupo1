**UF_CAMPUS/MUNICIPIO_CAMPUS**: Pode indicar a região geográfica e o potencial mercado consumidor para alimentos.

**NOME_CURSO**: Se relacionado a cursos de gastronomia, nutrição ou hotelaria, poderia indicar um interesse local em alimentos ou um mercado para restaurantes.

**TURNO**: Indica períodos de atividade do campus que podem influenciar o consumo em restaurantes ou mercados próximos.

**OPCAO**: Se refere à preferência do candidato, o que poderia ser correlacionado com os tipos de cursos mais procurados e, portanto, as tendências de consumo.

Outras variáveis como **GRAU**, **DS_PERIODICIDADE**, e **SEXO** podem oferecer informações contextuais adicionais, mas são menos diretamente relacionadas ao objetivo de identificar padrões de consumo.

```python
sns.countplot(x='TURNO', data=sisu)
plt.show()
```
